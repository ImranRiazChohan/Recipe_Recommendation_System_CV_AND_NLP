```python
import pandas as pd
```

```python
df=pd.read_csv('./input/RecipeDataset.csv')
```

```python
df.head()
```

```python
df['Ingredients_Clean'] = df['Ingredients'].str.replace(r'[', '')
df['Ingredients_Clean']=df['Ingredients_Clean'].str.replace(r']', '')
```

```python
df['Ingredients_Clean'][1]
```

```python
def add_quotes(text):
    quoted_items = ["'" + item.strip() + "'" for item in text.split(',')]
    return ', '.join(quoted_items)
df['Ingredients_Cleaned'] = df['Ingredients_Clean'].apply(add_quotes)
```

```python
df.head()
```

```python
# Now Save All Data into New CSV
new_df=df
new_df=new_df.drop(columns=['Ingredients','Ingredients_Clean'],axis=0)
new_df.head()
```

```python
new_df['Ingredients']=new_df['Ingredients_Cleaned']
new_df.head()
```

```python
new_df=new_df.drop(columns=['Ingredients_Cleaned'],axis=0)
new_df.head()
```

```python
# Function to add square brackets around a string
def add_square_brackets(text):
    return f"[{text}]"

# Apply the function to the 'Ingredients' column
new_df['Ingredients'] = new_df['Ingredients'].apply(add_square_brackets)
new_df.head()
```

```python
new_df.to_csv('indian_recipe_dataset.csv',index=False)
```
